```python
# calculate NDVI
import gdal
import numpy as np
import glob
import os

site = 'ersaa'
baseName = '/Volumes/RASMUS_1/Satellite/remains_sites'
inFile = glob.glob(os.path.join(baseName, site, 'sequoia') + '/*_seqirrad_empcorr.tif')[0]
outFile = inFile[:-4] + '_ndvi.tif'

def correct(array):
    array = np.where(array <= 0.00001, 0.00001, array)
    array = np.where(array > 1., 1., array)
    return array

ds = gdal.Open(inFile)
red = ds.GetRasterBand(2).ReadAsArray().astype(np.float32)#/32768.
nir = ds.GetRasterBand(4).ReadAsArray().astype(np.float32)#/32768.
red = correct(red)
nir = correct(nir)

ndvi = (nir - red) / (nir + red)

drv = gdal.GetDriverByName('GTiff')
outTif = drv.Create(outFile, ds.RasterXSize, ds.RasterYSize, 1, gdal.GDT_Float32)
outTif.SetGeoTransform(ds.GetGeoTransform())
outTif.SetProjection(ds.GetProjection())
outTif.GetRasterBand(1).WriteArray(ndvi)
outTif.GetRasterBand(1).SetNoDataValue(-999)
outTif = None
```

```python
# Calculate Biomass
import gdal
import numpy as np
import glob
import os

site = ''
baseName = '/Volumes/RASMUS_1/Satellite/remains_sites'
inFile = glob.glob(os.path.join(baseName, site, 'sequoia') + '/*_seqirrad_empcorr_ndvi.tif')[0]
outFile = inFile[:-4] + '_ndvi_biomass_hansen.tif'


ds = gdal.Open(inFile)
ndvi = ds.GetRasterBand(1).ReadAsArray().astype(np.float32)

# vores
#biomass = 0.0218*np.exp(3.7788*ndvi)
# Hansen
biomass = np.exp(5.7782*ndvi-3.194)/10


#=EXP($T$88*S65-$U$88)
drv = gdal.GetDriverByName('GTiff')
outTif = drv.Create(outFile, ds.RasterXSize, ds.RasterYSize, 1, gdal.GDT_Float32)
outTif.SetGeoTransform(ds.GetGeoTransform())
outTif.SetProjection(ds.GetProjection())
outTif.GetRasterBand(1).WriteArray(biomass)
outTif.GetRasterBand(1).SetNoDataValue(-999)
outTif = None
```
